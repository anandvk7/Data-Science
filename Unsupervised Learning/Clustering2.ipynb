```python
df = pd.read_csv('C:/Users/krishna/Downloads/unsupervised learning/circle.csv')
df.head()
```

```python
sns.pairplot(df)
```

# Appyling k means

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(n_clusters=2)
df['KM_Cluster'] = km.fit_predict(df[['Xcircle_X1','Xcircle_X2']])
```

visualize k mean cluster

```python
sns.scatterplot(x='Xcircle_X1', y='Xcircle_X2',data=df, hue='KM_Cluster')
```

# Applying DBSCAN

```python
from sklearn.cluster import DBSCAN 
```

```python
#Initialize DBSCAN
db = DBSCAN(eps=0.2,min_samples=10)
```

```python
#fit on the data
db.fit(df[['Xcircle_X1','Xcircle_X2']])
```

# how many clusters?

```python
np.unique(db.labels_)
```

```python
df['DB_Cluster'] = db.labels_
```

visualize clusters

```python
sns.scatterplot(x='Xcircle_X1', y='Xcircle_X2',data=df, hue='DB_Cluster')
```
