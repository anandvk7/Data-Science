```python
df[['bathrooms', 'stories', 'parking']] = df[['bathrooms', 'stories', 'parking']].astype(object)
```

```python
y = df['lprice']
X = df.drop(['price', 'lprice'], axis=1)
X = X.assign(const=1)
```

### Statistical Modeling

```python
import statsmodels.api as sm

ols = sm.OLS(y, X)
result = ols.fit()

result.summary()
```

```python
plt.figure(figsize=(15,10))
corr = X.corr()
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)]=True
sns.heatmap(corr, mask=mask, linewidth=0.5, annot=True)
plt.show()
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

pd.DataFrame({'vif':vif}, index=X.columns)
```

#### improving model accuracy by considering interaction and transformation

```python
from sklearn.preprocessing import PolynomialFeatures

X = X.drop('const', axis=1)

pf = PolynomialFeatures()
Xp = pf.fit_transform(X)
cols = pf.get_feature_names(X.columns)

Xp = pd.DataFrame(Xp, columns=cols)
colsxp = Xp.columns
```

```python
Xp.info(verbose=True)
```

```python
Xp.head()
```

```python
y.shape
```

```python
ys = y.values
ols = sm.OLS(ys, Xp)
result = ols.fit()

result.summary()
```

```python
## Backward Elimination to remove insignificant variables one by one

pmax=1
cols = list(Xp.columns)
X_1 = Xp.copy()

while (len(cols)>0):
    p = []
    X_1 = X_1[cols]
    model = sm.OLS(ys, X_1).fit()
    p = model.pvalues
    pmax = max(p)
    
    if (pmax>0.05):
        cols = list(p.drop(p[p==pmax].index).index)
        #print(cols)
    else:
        break
        
selected = cols
print(selected)
print(len(selected))
```

```python
Xs = Xp.loc[:,selected]

ols = sm.OLS(ys, Xs)
result = ols.fit()

result.summary()
```
