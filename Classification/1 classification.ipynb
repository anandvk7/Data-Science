# Logistic Regression_Statistics Approach

```python
import statsmodels.api as sm

y = df['TenYearCHD']
x = df.drop('TenYearCHD',axis=1)
```

```python
x_const = sm.add_constant(x)
```

```python
model = sm.Logit(y,x_const)
result = model.fit()
result.summary()
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = [variance_inflation_factor(x.values,i) for i in range(x.shape[1])]
pd.DataFrame({'vif':vif}, index = x.columns)
```

```python
## Backward elimantion to drop insignificant variables one by one

cols = list(x_const.columns)
p=[]
while len(cols)>1:
    x_const = x_const[cols]
    model = sm.Logit(y,x_const).fit().pvalues
    p = pd.Series(model.values[1:],index=x_const.columns[1:])
    pmax = max(p)
    pid = p.idxmax()
    if pmax>0.05:
        cols.remove(pid)
        print('Variable removed:',pid,pmax)
    else:
        break
cols
```

```python
model = sm.Logit(y,x_const)
result = model.fit()
result.summary()
```

```python
df.to_csv('chd.csv')
```

```python
np.exp(result.params)
```

```python
prob = result.predict(x_const[cols])
```

```python
prob
```

```python
prob = pd.DataFrame(prob,columns=['prob'])
prob['y_est'] = prob['prob'].apply(lambda x: 1 if x>=0.5 else 0)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report

confusion_matrix(y, prob['y_est'])
```

```python
accuracy_score(y, prob['y_est'])
```

```python
print(classification_report(y, prob['y_est']))
```

# ROC Analysis

```python
from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression(solver='liblinear', fit_intercept=True)

logreg.fit(X_train, y_train)

y_prob_train = logreg.predict_proba(X_train)[:,1]
y_pred_train = logreg.predict(X_train)


y_prob = logreg.predict_proba(X_test)[:,1]
y_pred = logreg.predict(X_test)
```

```python
from sklearn.metrics import roc_auc_score, roc_curve

print('AUC of train: ',roc_auc_score(y_train, y_prob_train))

fpr, tpr, threshold = roc_curve(y_train,y_prob_train)

plt.plot(fpr, tpr)
plt.plot(fpr,fpr,'r-')
plt.xlabel('fpr')
plt.ylabel('tpr')
plt.title('ROC of train data')
plt.show()
```

```python
print('AUC of train: ',roc_auc_score(y_test, y_prob))

fpr, tpr, thresholds = roc_curve(y_test,y_prob)

plt.plot(fpr, tpr)
plt.plot(fpr,fpr,'r-')
plt.xlabel('fpr')
plt.ylabel('tpr')
plt.title('ROC of test data')
plt.show()
```
